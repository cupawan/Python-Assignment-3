```python
def myreduce(f, v):
    z = iter(v)
    i = next(z)
           
        
    a = i
    for x in z:
        a = f(a, x)
    return a

#demonstration of  myreduce() function by taking a lambda function which returns sum of two elements as "f" and a list as "v" :

myreduce(lambda p,q: p+q ,[1,2,3,4,5])           
    
```

```python
# Another demonstration of myreduce() function
myreduce(lambda p,q : p*q, [1,2,3,5])
```

```python
def myfilter(function,iterable):
    res = []
    for i in iterable:
        if function(i) == True:
            res.append(i)
        else:
            continue
    return res
```

```python
# Demonstration of myfilter() function:
l = [1,2,3,45,33,29,30,22,200,345,44]
myfilter(lambda x : True if x>30 else False,l)
```
